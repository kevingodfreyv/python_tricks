### Series
A Series is like a single column of data—just one axis.
Think of it as a simple list or array

Unlike a plain list or a NumPy array, each value in a Series has a label, called an index

This labeling allows you to access data by name instead of just position:

By position → series[0]
By label → series['a']

```python
# Series - 1D labeled array
series = pd.Series([10, 20, 30, 40, 50])
print("Series with default index:")
print(series)
print(series[3])

```

```python

# Series with custom index
series_custom = pd.Series([10, 20, 30, 40, 50], index=[44, 33, 'c', 'd', 'e'])
print("\nSeries with custom index:") # once custom index is assigned, default index get overwritten.
print(series_custom)
print(series_custom[33], series_custom['c'])
```

```python
# Series from dictionary
series_dict = pd.Series({'apple': 5, 'banana': 10, 'orange': 7, 'grape': 12})
print("\nSeries from dictionary:")
print(series_dict)
print(series_dict['banana'])
print(series_dict.iloc[1]) # even after assigning custom index, we can access values via position using iloc.
```

```python
# Series operations
series = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

print("Original series:")
print(series)

print("\nAccess by label (series['b']):", series['b'])
# print("Access by position (series[1]):", series[1])

print("\nSeries + 5:")
print(series + 5)

print("\nSeries * 2:")
print(series * 2)

print("\nSeries properties:")
print("Values:", series.values)
print("Index:", series.index)
print("Name:", series.name)
```

# Question: casting columns to suitable types

```python
# type conversions
duplicated_df["student"] = duplicated_df["student"].astype("category")
duplicated_df["language"] = duplicated_df["language"].astype("float32")
duplicated_df["maths"] = duplicated_df["maths"].astype("int32")
duplicated_df.dtypes
# type casting a object/string into category is a huge win.
# category, stores only n unique strings and their respective integer encodings
# and replaces these string values with integers behind the scenes.
# this is much more memory efficient than storing strings.
# operations such as Grouping , filtering becomes faster
# Category => Encoded + optimized representation.
```

# Perform lower, upper, contains and starts with operations on Series Object

```python
# String operations
df_string = pd.DataFrame({"student":["Kevin014", "Kiran", "Ritesh", "Suresh", "Vignesh","esh_esh_esh", "Rishkesh55", "vis4leshish", "veshnish"]})

# String methods
print("\nLowercase:")
print(df_string['student'].str.lower().head())

print("\nUppercase:")
print(df_string['student'].str.upper().head())

print("\nContains 'esh':")
print(df_string['student'].str.contains('esh').head())

print("\nStartswith 'k':")
print(df_string['student'].str.lower().str.startswith('k').sum(), "students")

```

```python
# applying a function for all the values in a column
def grader(score):
    grade = "F"
    if(score >= 90):
        grade = "A"
    elif(score >= 80):
        grade = "B"
    elif(score >= 70):
        grade = "C"
    return grade

# but this way of applying is slower.
df['grade'] = df['science'].apply(grader)
df
```
